```python
import pandas as pd
import numpy as np

df = pd.read_parquet('turbo_df.parquet')
df.head(20)
```

```python
df.info()
```

```python
df['Classified Date'] = pd.to_datetime(df['Classified Date'])
df['Owners count'].value_counts()
```

```python
df["Seats count"] = pd.to_numeric(
    df["Seats count"].replace("8+", 8),
    errors="coerce"
).astype("Int64")
df["Owners count"] = pd.to_numeric(
    df["Owners count"].replace("4 və daha çox", 4),
    errors="coerce"
).astype("Int64")
```

```python
df.info()
```

```python
df.head()
```

```python
df["Seats count"] = df["Seats count"].fillna(df["Seats count"].median()).astype("Int64")
df["Owners count"] = df["Owners count"].fillna(df["Owners count"].median()).astype("Int64")
df["Condition"] = df["Condition"].fillna(df["Condition"].mode()[0])
df["Supply"] = df["Supply"].fillna(df["Supply"].mode()[0])

df.isna().sum()
```

```python
azn_rates = {"AZN": 1.0, "USD": 1.70, "EUR": 1.95}
```

```python
df["Price (AZN)"] = (df["Price"] * df["Currency"].map(azn_rates)).round(2)
df = df.drop(columns=["Price", "Currency", "Description"])
df.head()
```
